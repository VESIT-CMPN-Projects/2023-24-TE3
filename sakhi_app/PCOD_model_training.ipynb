```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('/content/sample_data/PCOS_data_without_infertility_cp.xlsx')
df
```

```python
df.drop(['IIbeta-HCG(mIU/mL)'],axis=1,inplace=True)
```

```python
df.drop(['Unnamed: 44'],axis=1,inplace=True)
df.drop(['AMH(ng/mL)'],axis=1,inplace=True)
```

```python
df
```

```python
X=df.drop(['Sl. No', 'Patient File No.', 'PCOS (Y/N)'],axis=1)
y=df['PCOS (Y/N)']
```

```python
print(df.dtypes)
```

```python
X
```

```python
X.isnull().sum()
```

```python
X['Fast food (Y/N)'].unique()
```

```python
# df['AMH(ng/mL)'].replace('a',5.62, inplace=True)
X['Marraige Status (Yrs)'].replace(np.nan,7.68,inplace=True)
X['Fast food (Y/N)'].replace(np.nan,1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy Score ",accuracy_score(y_test,y_pred)*100)
```

```python
import joblib
joblib.dump(clf,'PCOD-predictions.joblib')
```

```python
clf=joblib.load('PCOD-predictions.joblib')
predictions=clf.predict([[36,52,150,23.111111,15,80,20,10,4,2,4,0,0,1.99,2.8,1.51,
                          1.854306,40,38,0.95,6.65,11.74,27.7,0.25,125,0,0,0,0,0,0,0,
                          110,80,1,1,14,17,2.5]])
predictions
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
svm=SVC()
svm.fit(X_train,y_train)
y_preds=svm.predict(X_test)
print("Accuracy Score ",accuracy_score(y_test,y_preds)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 1000, random_state = 42)
rf.fit(X_train, y_train)
y_predsrf = rf.predict(X_test)
print("Accuracy Score ",accuracy_score(y_test,y_predsrf)*100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_preds = knn.predict(X_test)
print("Accuracy Score ",accuracy_score(y_test,y_preds)*100)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
logisticRegr = LogisticRegression(solver='lbfgs', max_iter=10000)
logisticRegr.fit(X_train, y_train)
predictions = logisticRegr.predict(X_test)
print("Accuracy Score ",accuracy_score(y_test,predictions))
print(classification_report(y_test,predictions))
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)
predictednb = model.predict(X_test)
print("Accuracy Score ",accuracy_score(y_test,predictednb)*100)
```

```python
from sklearn.ensemble import AdaBoostClassifier
abc = AdaBoostClassifier(n_estimators=50,
                         learning_rate=1)
model = abc.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Accuracy Score ",accuracy_score(y_test,y_pred)*100)
```
